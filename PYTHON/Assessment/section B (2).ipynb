```python
# 1. Load the sales_data_sample.csv file and use Python string methods to trim 
# whitespace and apply title casing to the 'PRODUCTLINE' column. 

import pandas as pd

df = pd.read_csv(
    r"D:\Data Analyst\PYTHON\Assessment\sales_data_sample.csv",
    encoding="latin1"
)

print(df.head())
```

```python
df["PRODUCTLINE"] = df["PRODUCTLINE"].str.strip().str.title()
print(df["PRODUCTLINE"].head())
```

```python
# 2. Iterate through the 'SALES' column to remove non-numeric characters and 
# cast the cleaned values into a float data type. 

df["SALES"] = df["SALES"].astype(str)

df["SALES"] = df["SALES"].str.replace(r"[^0-9.]", "", regex=True)

df["SALES"] = df["SALES"].astype(float)

print(df["SALES"].head())
```

```python
# 3. Construct a dictionary where each key is a unique 'PRODUCTLINE' and the value 
# is the cumulative sum of 'SALES' for that category. 
sales_dict = {}

for i in range(len(df)):
    product = df["PRODUCTLINE"][i]
    sale = df["SALES"][i]

    if product in sales_dict:
        sales_dict[product] += sale
    else:
        sales_dict[product] = sale

print(sales_dict)
```

```python
# 4. Use a loop and conditional logic to categorize product lines as "High Revenue 
# Product" if total sales exceed 100,000, printing the results to the console. 

for product, total in sales_dict.items():

    if total > 100000:
        print(product, "- High Revenue Product")
    else:
        print(product, "- Standard Product")
```
