```python
# 4. Compare using a manual for loop versus the sum() function, identifying a 
# scenario where the loop is the only viable option. 

# Using sum():
numbers = [10, 20, 30]
print(sum(numbers))

# Using a loop:

total = 0
for num in numbers:
    if num > 15:
        total += num

print(total)

# When loop is required:
# When conditions or filtering are needed before adding values.
```

```python
# 5. Describe how to implement type checking or error handling to prevent a 
# TypeError when comparing a potential string variable to a number. 

# Use type checking or exception handling.
value = "100"
if isinstance(value, (int, float)):
    print(value > 50)
else:
    print("Value is not numeric")



try:
    value = float(value)
    print(value > 50)
except ValueError:
    print("Invalid number")
```
