```python
# 1.
# Create a Python script that takes any product name string 
# (e.g., 'Redmi Note 12 Pro') and prints the name in all uppercase and 
# all lowercase using the upper() and lower() methods.

product_name = "Redmi Note 12 Pro"

print("original :- ", product_name)
print("Uppercase :- ", product_name.upper())
print("lowercase :- ", product_name.lower())
```

```python
# 2.
# Write a function clean_brand_name(name) that removes leading/trailing spaces 
# and replaces any hyphens '-' with a single space in the input string. 
# Test it with ' oneplus-Nord '.

def clean_brand_name(name):
    return name.strip().replace("-", " ")

brand = "oneplus-Nord"
print(clean_brand_name(brand))
```

```python
# 3.
# Given the string 'Apple iPhone 14 Pro Max', use string slicing to extract and 
# print only the brand name and the model (i.e., 'Apple' and 'iPhone 14 Pro Max')
#  separately.<br><br><em><strong>Hint:</strong> Use split() to help find the 
# split point, then use slicing for the substrings.</em>

x = "Apple iPhone 14 Pro Max"

split_index = x.find(" ")

brand = x[:split_index]
model = x[split_index + 1:]

print("Brand :- ", brand)
print("Model :-",model)


```

```python
# 5.
# Suppose you have a list of messy product names: [' mi-Band 5 ', ' SAMSUNG-Galaxy ', ' realme-Book ']. 
# Write code to clean each name (remove spaces, replace hyphens with spaces, and make the brand title case) 
# and print the cleaned list.<br><br><em><strong>Constraint:</strong> Use at least three string methods from
#  this session.</em>

product = [' mi-Band 5 ', ' SAMSUNG-Galaxy ', ' realme-Book ']

cleaned_product = []

for product in product:
    cleaned = product.strip()        # Remove leading/trailing spaces
    cleaned = cleaned.replace("-", " ")  # Replace hyphen with space
    cleaned = cleaned.title()        # Convert to title case
    cleaned_product.append(cleaned)

print(cleaned_product)

```
