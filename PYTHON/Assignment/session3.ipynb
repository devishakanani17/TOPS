```python
# 1.
# Declare four variables in Python: one integer (number of followers), 
# one float (average rating), one string (your favorite app's name), and 
# one boolean (is_premium_user). Print each variable and its type using the
#  type() function.

followers = 5000
average_rating = 4.8
favorite_app = "Instagram"
is_premium_user = True

print(followers, type(followers))
print(average_rating, type(average_rating))
print(favorite_app, type(favorite_app))
print(is_premium_user, type(is_premium_user))
```

```python
# 2.
# Write a Python program that takes a user's input for the price of a Zomato order as a string, 
# converts it to a float using type casting, adds 18% GST, and prints the final bill amount.

price = (input("Enter Zomato order price :- "))

price = float(price)
final_bill = price + (price * 18 / 100)

print("final Bill:", final_bill)

```

```python
# 3.
# Given a list of strings representing product prices from Flipkart, like ['199.99', '299.50', '150'], 
# convert all to floats and calculate the total cart value.

price = ['199.99', '299.50', '150']

total = 0

for price in price:
    total += float(price)

print("the total cart value:- ", total)
```

```python
# 4.
# Build a function is_discount_applicable(order_amount) that takes a float and returns True if the amount is
# greater than 500, otherwise False. Print the result for order amounts 450 and 750.

def is_discount_applicable(order_amount):
    return (order_amount > 500)

print(is_discount_applicable (450))
print(is_discount_applicable(750))
```

```python
# 5.
# You received a dataset of ratings as strings from Spotify: ['4.5', '3.0', '5', '4.2']. Use type casting to 
# convert these to floats, then find and print the highest rating.<br><br><em><strong>Hint:</strong> Use 
# the float() function inside a loop or list comprehension.</em>

ratings = ['4.5', '3.0', '5', '4.2']

float_ratings = []
for rating in ratings:
    float_ratings.append(float(rating))

print("Highest Rating",max(float_ratings))
```
