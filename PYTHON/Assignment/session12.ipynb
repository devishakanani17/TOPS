```python
# 1.
# Write a lambda function that takes a price in rupees and returns the price after adding 18% GST. 
# Test it on the prices 100, 250, and 500.

def gst(price):
    return price * 1.18

print(gst(100))
print(gst(250))
print(gst(500))
```

```python
# 2.
# Given a list of song titles from Spotify with extra spaces and inconsistent casing, use map() and a lambda 
# function to clean each title so that it is stripped of spaces and converted to title case (e.g., ' shape OF you ' → 'Shape Of You').

songs = [" shape OF you ", " believer ", " blinding LIGHTS "]

clean_songs = list(map(lambda song: song.strip().title(), songs))

print(clean_songs)
```

```python
# 3.
# Use filter() and a lambda function to extract only those Flipkart product names from a list that 
# start with the letter 'S' (case-insensitive).

products = ["Shoes", "shirt", "Socks", "Bag", "speaker"]

result = list(filter(lambda x: x.lower().startswith("s"), products))

print(result)
```

```python
# 4.
# Given a list of order amounts from a Zomato cart [120, 340, 560, 80], use reduce() from functools to
#  calculate the total bill amount.

from functools import reduce

orders = [120, 340, 560, 80]
total = reduce(lambda x, y: x + y, orders)

print(total)
```

```python
# 5.
# Use ChatGPT or Copilot to generate a Python code snippet that uses map(), filter(), and reduce() together to 
# process a list of numbers: first double each number, then filter to keep only numbers greater than 100, and 
# finally sum the result. Paste and test the generated code with the list [40, 60, 80, 120].

from functools import reduce

num = [40, 60, 80, 120]

doubled = map(lambda x: x * 2, num)
filtered = filter(lambda x: x > 100, doubled)
result = reduce(lambda x, y: x + y, filtered)

print(result)


# Step-by-step:
# Original list: [40, 60, 80, 120]
# After doubling: [80, 120, 160, 240]
# Keep numbers greater than 100: [120, 160, 240]
# Sum: 120 + 160 + 240 = 520
```
