```python
# 1.
# Use re.findall() to extract all valid Indian phone numbers (10 digits, starting with 7, 8, or 9) from a given
#  text string that contains random numbers, prices, and phone numbers like those seen in OLX or WhatsApp chats.

import re

text = "Call me at 9876543210 or 8123456789. Price is 500. Wrong number: 6123456789"

phones = re.findall(r"[789]\d{9}", text)

print(phones)
```

```python
# 2.
# Write a Python function using re.search() that checks if a given string contains a valid date in the format
#  DD/MM/YYYY (e.g., 25/06/2024), and returns True if found, otherwise False.<br><br><em><strong>Hint:</strong> 
# Use the pattern '\b\d{2}/\d{2}/\d{4}\b'.</em>

import re

def check_date(text):
    if re.search(r"\b\d{2}/\d{2}/\d{4}\b", text):
        return True
    return False

print(check_date("Today's date is 25/06/2024"))
print(check_date("No date here"))
```

```python
# 3.
# Given a messy text copied from a Zomato review containing multiple emails, use re.findall() to extract 
# all valid email addresses and print them as a list.

import re

text = "Contact us at support@gmail.com or help@yahoo.com"

emails = re.findall(r"\S+@\S+\.\S+", text)

print(emails)
```

```python
# 4.
# Use re.sub() to mask all but the last 4 digits of any phone number in a string 
# (e.g., replace 9876543210 with ******3210) like Paytm does for privacy.<br><br><em><strong>Constraint:</strong>
#  Do not use loops; achieve this only with re.sub().</em>

import re

text = "My number is 9876543210"

result = re.sub(r"\d{6}(\d{4})", r"******\1", text)

print(result)
```

```python
# 5.
# Use ChatGPT to generate a regex pattern that matches Flipkart-style order IDs (e.g., OD123456789012345000)
#  and test it in Python using re.search() on sample order strings.

import re

order_id = "OD123456789012345000"

pattern = r"OD\d{18}"

if re.search(pattern, order_id):
    print("Valid Order ID")
else:
    print("Invalid Order ID")
```
