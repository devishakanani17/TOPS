```python
# 3.
# Write a function remove_last_item(order_list) that pops the last item from a
# Zomato order list and returns the removed item. Test it with a sample order_list.

def remove_last_item(order_list):
    return order_list.pop()

order = ["Pizza", "Cheese cake", "mocktaill", "Brownie"]

removed_item = remove_last_item(order)

print("Removed Item:", removed_item)
print("Remaining Order:", order)
```
